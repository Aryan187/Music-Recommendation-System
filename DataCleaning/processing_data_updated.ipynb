```python
#takes a lot of time: run only if required, output already in 'cleaned_listofsongs_and_behavior_peruser.csv'
user_songs={}
for i in dataframe.index:
    if dataframe['msno'][i] not in users:
        continue
    if dataframe['song_id'][i] in songs:
        if dataframe['msno'][i] not in user_songs:
            user_songs[dataframe['msno'][i]]=[]
        user_songs[dataframe['msno'][i]].append((dataframe['song_id'][i],dataframe['source_system_tab'][i]))
```
