```python
dataframe=pd.read_csv('test.csv')
dataframe.head()
```

```python
dataframe=dataframe.drop(['source_system_tab','source_screen_name','source_type'],axis=1)
dataframe=dataframe.dropna(subset=['msno','song_id'])
```

```python
user_songs={}
for i in dataframe.index:
    if dataframe['msno'][i] not in user_songs:
        user_songs[dataframe['msno'][i]]=[]
    user_songs[dataframe['msno'][i]].append(dataframe['song_id'][i])
```

```python
dataframe.head()
```

```python
import csv
```

```python
with open('dict.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for key, value in user_songs.items():
       writer.writerow([key, value])
```
