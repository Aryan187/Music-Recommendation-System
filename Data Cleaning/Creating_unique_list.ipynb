```python
import numpy as np
import pandas as pd
songs=pd.read_csv('final_songs.csv')
```

```python
songs.head()
```

```python
artistlist=songs['artist_name']
b=[]
for i,j in enumerate(artistlist):
    x=j[1:-1].split(', ')
    b.insert(i,x)
blist = [item for sublist in b for item in sublist]
artist=np.unique(np.array(blist))
pd.DataFrame(artist).to_csv('Artist.csv')
```

```python
composerlist=songs['composer']
b=[]
for i,j in enumerate(composerlist):
    x=j[1:-1].split(', ')
    b.insert(i,x)
blist = [item for sublist in b for item in sublist]
composer=np.unique(np.array(blist))
pd.DataFrame(composer).to_csv('Composer.csv')
```

```python
genrelist=songs['genre_ids']
b=[]
for i,j in enumerate(genrelist):
    x=j[1:-1].split(', ')
    b.insert(i,x)
blist = [item for sublist in b for item in sublist]
genre=np.unique(np.array(blist))
pd.DataFrame(genre).to_csv('Genre.csv')
```

```python
pd.DataFrame(np.unique(np.array(songs['language']))).to_csv('Language.csv')
```
